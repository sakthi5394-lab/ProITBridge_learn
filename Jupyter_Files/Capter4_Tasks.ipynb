###  1.	Write a program to take a name and age from the user and print a greeting message.

```python
try:
    user_name = input("Enter The Name:")
    user_age  = int(input("Enter the Age:"))
    print(f"Hey! Welcom Mr/Mrs . {user_name}, Your age is {user_age}")
except Exception as e:
    print(f"Please validate the Error {e}")
finally:
    print("Thanks! Welcome Back!!")
```

### 2.	Create a list of numbers and print only even numbers using a for loop.

### 3.	Write a program that handles division by zero using try-except.

```python
def division():
    try:
        num = int(input("Enter the numarator: "))
        dem = int(input("Enter the denominator: "))
        result = num/dem
        return result
    except Exception as e:
        print(f"Enter the valid numbers other than 0: {e} error")
    finally:
        print("Thanks!!!")

print(division())
```

###  4.	Given a string, count how many vowels it has.

```python
user_input = input("Enter the string: ")
count:int = 0
for i in user_input.strip().upper():
    if i in "AEIOU":
        count += 1
print(f"Hey! Total Number {count} vowels present in the string' {user_input}'")
```

### 5.	Reverse a list without using built-in functions.

```python
user_list = [1,2,3,4,5,10,6,97]
lst_len = len(user_list)
reverse_list = [0]*lst_len
start = 0
while start < lst_len:
    reverse_list[start] = user_list[lst_len-1]
    reverse_list[lst_len-1] = user_list[start]
    start += 1
    lst_len -= 1

reverse_list
    
```

## Task 4
Lambda Function 2 examples with Map, Filter, Reduce


```python
import pandas as pd
euro_game = pd.read_csv("Data_File/Pandas_CrassCourse/Euro_2012_stats_TEAM.csv")
euro_game
```

```python
euro_game.nunique()
```

```python
euro_game.head()
```

```python
euro_game = euro_game[["Team","Goals","Shots on target","Total shots (inc. Blocked)","Fouls Won","Yellow Cards","Red Cards","Players Used"]]
```

```python
euro_game.sort_values(by = "Goals",ascending= False)
```

```python
avg_Rate = euro_game["Goals"].mean().round().astype(int)
avg_Rate
```

```python
euro_game.loc[:,"Team_Strength"] = euro_game["Goals"].map(lambda x: "above Average" if int(x)> avg_Rate else "below_average") 
```

```python
avg_shots = euro_game["Shots on target"].mean().round().astype(int)
avg_shots
```

```python
euro_game.sort_values(by = "Goals",ascending= False)
```

```python
euro_game.loc[:,"Team_Strngth_Based_On_Shots"] = euro_game["Shots on target"].map(lambda x: "Very Strong Team" if x > avg_shots else "Average Team" ) 
```

```python
euro_game["Shots on target"].map(lambda x: "Very Strong Team" if x > avg_shots else "Average Team" ) 
```

```python
euro_game.sort_values(by = "Goals")
```

#### Filter Function

```python
euro_game.filter(items= ["Team","Goals","Team_Strength"])

euro_game[filter(lambda col: col.startswith('Team') , euro_game.columns)]
```

```python
euro_game[filter(lambda col: col in ["Team","Goals","Team_Strength"] , euro_game.columns)]
```

#### Reduce 

```python
from functools import reduce
```

```python
## Total Goals in the Tournamends 
reduce(lambda x,y:x+y,euro_game["Goals"])
```

```python
### Total Yellow Cards in the tournament
reduce(lambda x,y:x+y,euro_game["Yellow Cards"])
```
