## 2. Another Method to find the recvommendation using Equdian distance

```python
from sklearn.metrics import pairwise_distances
```

```python
##Pivort the Data As per Movies 

EQ_mv_pivot_Data = pd.pivot_table(data = mv_data,values="rating",index="userId",columns="movie").fillna(0)
EQ_mv_pivot_Data.columns = mv_data["movie"].unique()
EQ_mv_pivot_Data
```

```python
mv_Equidian_Corr = pd.DataFrame(pairwise_distances(X = EQ_mv_pivot_Data,metric='euclidean'),columns=mv_data["userId"].unique(),index=mv_data["userId"].unique())
mv_Equidian_Corr
```

### 2.1 Cosine metric

```python
mv_cosine_Corr = pd.DataFrame(1-pairwise_distances(X = EQ_mv_pivot_Data,metric='cosine'),columns=mv_data["userId"].unique(),index=mv_data["userId"].unique())
mv_cosine_Corr
```
