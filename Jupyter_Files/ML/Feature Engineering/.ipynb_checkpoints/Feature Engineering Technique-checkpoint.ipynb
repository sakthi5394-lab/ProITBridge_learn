## 1. Import Necessary Library

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.feature_selection import RFE,SelectFromModel
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
import warnings

warnings.filterwarnings("ignore")
```

## 2. Data Collection

```python
cancer_data = load_breast_cancer()
```

```python
Cancer_Data_Df = pd.DataFrame(cancer_data.data,columns=cancer_data.feature_names)
Cancer_Data_Df["Target"] = cancer_data.target
```

```python
Cancer_Data_Df
```

## 3. Data Understanding 

```python
Cancer_Data_Df.shape
```

## 4. Module Building

```python
X = Cancer_Data_Df.drop("Target",axis = 1)
y = Cancer_Data_Df.Target
```

```python
X.shape,y.shape
```

## 5. Module Training || 6 Module Teating || 7. Module Evaluation

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,shuffle=True,random_state=12,stratify=y)
```

```python
def Random_Module_Exe(X_train,X_test,y_train,y_test): 
    rf_Module = RandomForestClassifier(random_state = 12)
    rf_Module.fit(X_train,y_train)
    y_predic_Test = rf_Module.predict(X_test)
    
    print("Accusracy Score : ",round(accuracy_score(y_test,y_predic_Test),4))
    print("Precision Score : ",round(precision_score(y_test,y_predic_Test),4))
    print("Recall Score : ",round(recall_score(y_test,y_predic_Test),4))
    print("Confusion Matrix : \n",confusion_matrix(y_test,y_predic_Test))

Random_Module_Exe(X_train,X_test,y_train,y_test)
```

# How to use the Feature Engineerting

## 1. SelectForm Module

```python
sf_module = SelectFromModel(estimator = RandomForestClassifier(random_state = 12),max_features=None)
```

```python
sf_module.fit(X_train,y_train)
```

```python
X.columns[sf_module.get_support()]
```

```python
len(X.columns[sf_module.get_support()])
```

```python
X_train_fe = sf_module.transform(X_train)
X_test_fe = sf_module.transform(X_test)
X_train_fe.shape,X_test_fe.shape
```

```python
%%time
Random_Module_Exe(X_train_fe,X_test_fe,y_train,y_test)
```

## 2. RFE Module

```python
rfe_Module = RFE(estimator = RandomForestClassifier(random_state = 12),n_features_to_select = 12)
```

```python
rfe_Module.fit(X_train,y_train)
```

```python
X_train_rfe = rfe_Module.transform(X_train)
X_test_rfe = rfe_Module.transform(X_test)
X_train_rfe.shape,X_test_rfe.shape
```

```python
%%time
Random_Module_Exe(X_train_rfe,X_test_rfe,y_train,y_test)
```

```python
%%time
Random_Module_Exe(X_train,X_test,y_train,y_test)
```

## Try to Use the Gradient Boosting Technique for select Feature asnd pass the value into Randomforest

```python
rfe_Module = RFE(estimator = GradientBoostingClassifier(random_state = 12),n_features_to_select = 12)
rfe_Module.fit(X_train,y_train)
X_train_rfe = rfe_Module.transform(X_train)
X_test_rfe = rfe_Module.transform(X_test)
X_train_rfe.shape,X_test_rfe.shape
```

```python
%%time
Random_Module_Exe(X_train_rfe,X_test_rfe,y_train,y_test)
```

## How to get the Optimal Feature Value to Select 

```python
for i in range(1,16):
    rfe_Module = RFE(estimator = GradientBoostingClassifier(random_state = 12),n_features_to_select = i)
    rfe_Module.fit(X_train,y_train)
    X_train_rfe = rfe_Module.transform(X_train)
    X_test_rfe = rfe_Module.transform(X_test)
    print("Select Fesute Count: ", i)
    Random_Module_Exe(X_train_rfe,X_test_rfe,y_train,y_test)
    print("##############################################################################################")
```
