### 1. Import Nessary Library

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
from sklearn.datasets import load_breast_cancer
from datetime import datetime
```

### 2. Data Preparation

```python
breast_cacncer_data = load_breast_cancer()
```

```python
breast_cacncer_df = pd.DataFrame(breast_cacncer_data.data,columns= breast_cacncer_data.feature_names)
breast_cacncer_df["Target"] = breast_cacncer_data.target
breast_cacncer_df
```

## 4. Data Preparation

```python
X = breast_cacncer_df.drop("Target",axis= 1)
y = breast_cacncer_df.Target
```

```python
y
```

```python
scler = StandardScaler()
```

```python
scaled_X = scler.fit_transform(X)
```

### 5. Module Buiding

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(scaled_X,y,train_size=0.8,random_state=12,shuffle=True,stratify=y)
```

```python
start_time = datetime.now()
gba_clasifier = GradientBoostingClassifier(n_estimators=50,max_depth=3)
End_Time = datetime.now()
print("Module Training Time : ", End_Time-start_time)
```

### 6. Module Training

```python
start_time = datetime.now()
gba_clasifier.fit(Xtrain,ytrain)
End_Time = datetime.now()
print("Module Training Time : ", End_Time-start_time)
```

## 7. Module Testing || 8 Module Ebavulation

```python
ytrain_predict = gba_clasifier.predict(Xtrain)
```

```python
print("Confusion Matrix : \n",confusion_matrix(ytrain,ytrain_predict))
print("Accuracy Score :",accuracy_score(ytrain,ytrain_predict))
print("Precision Score :",precision_score(ytrain,ytrain_predict))
print("Recall_Score :",recall_score(ytrain,ytrain_predict))
```

```python
ytest_predict = gba_clasifier.predict(Xtest)
```

```python
print("Confusion Matrix : \n",confusion_matrix(ytest,ytest_predict))
print("Accuracy Score :",accuracy_score(ytest,ytest_predict))
print("Precision Score :",precision_score(ytest,ytest_predict))
print("Recall_Score :",recall_score(ytest,ytest_predict))
```
