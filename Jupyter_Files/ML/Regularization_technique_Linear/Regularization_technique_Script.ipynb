```python
##1. Import Library
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression,ElasticNet,Ridge,Lasso
from sklearn.model_selection import train_test_split
## 2. Load the Data
housing_data = fetch_california_housing()
housing_data
housing_DF = pd.DataFrame(housing_data.data,columns=housing_data.feature_names)
housing_DF["Target"] = housing_data.target
housing_DF
X = housing_DF.drop("Target",axis= 1)
y = housing_DF.Target
print(X.shape,y.shape)

## Take only 350 Observation
X = X.head(350)
y = y[:350]

## Module Building

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size= 0.80,shuffle=True,random_state=42)
ln_module = LinearRegression()
rg_module = Ridge(alpha=1.0)
ls_Module = Lasso(alpha=0.1)
es_module = ElasticNet(alpha=0.1)



## Module Train
ln_module.fit(X_train,y_train)
rg_module.fit(X_train,y_train)
ls_Module.fit(X_train,y_train)
es_module.fit(X_train,y_train)

print("Linear_Coeff:",ln_module.coef_)
print("Gamma_Coeff:",rg_module.coef_)
print("Lasso_Coeff:",ls_Module.coef_)
print("Elastic_Coeff:",es_module.coef_)
```

```python
#### Displayt the Error

from sklearn.metrics import mean_squared_error,r2_score

module = {
        "Linear_Module":ln_module,
        "Ridge_Module":rg_module,
        "Lesso_Module": ls_Module,
        "Elastic_Module": es_module
        }

for name,module in module.items():
    y_predict_test = module.predict(X_test)
    print(name)
    print("MAE Value :",round(mean_squared_error(y_test,y_predict_test),2))
    print("R2_Score :", round(r2_score(y_test,y_predict_test),2))
    print()
```
