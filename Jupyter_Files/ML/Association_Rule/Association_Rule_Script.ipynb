## Data Mining

```python
uk_Online_Data.dropna(inplace = True)
```

```python
uk_Online_Data.isna().sum()
```

```python
uk_pivort_Data = uk_Online_Data.pivot_table(values="Quantity",index="InvoiceNo",columns="Description",aggfunc = "sum").fillna(0)
```

```python
uk_pivort_Data_onehot = uk_pivort_Data.applymap(lambda x: 0 if x == 0 else 1)
```

```python
## Online Data One Hotted VAlue Data Frame
uk_pivort_Data_onehot
```

## ML Exdent Association

```python
apprior_DF = apriori(df = uk_pivort_Data_onehot ,min_support=0.02,use_colnames = True)
```

```python
best_Associates = association_rules(df = apprior_DF,min_threshold = 0.2)
```

```python
best_Associates
```

## Export to CSV

```python
best_Associates.to_csv("Best_Association_List.csv")
```
