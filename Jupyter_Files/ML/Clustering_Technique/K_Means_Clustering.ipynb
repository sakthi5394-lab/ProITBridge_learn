## 1. Import Necessary Library

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

## 2. Data Collection 

```python
collegae_dataset = pd.read_csv("college_list.csv")
collegae_dataset
```

## 3. Data Preparation

```python
X = collegae_dataset.drop("College_Name",axis= 1)
```

```python
X
```

```python
scaller_mdl = StandardScaler()
```

```python
scalled_X = scaller_mdl.fit_transform(X)
```

```python
scalled_X = pd.DataFrame(scalled_X,columns=X.columns)
```

```python
scalled_X.mean().round(),scalled_X.std().round()
```

## 4.Clustering

```python
K_modle = KMeans(n_clusters = 3)
```

```python
K_modle.fit_predict(scalled_X)
```

```python
K_modle.n_clusters
```

```python
collegae_dataset["College_Category"] = K_modle.labels_
```

## Find the Optimal Cluster NUmber

```python
## WCSS (With in cluster Square)
wcss = []
for i in range(1,11):
    k_module = KMeans(n_clusters = i,random_state = 12)
    k_module.fit(scalled_X)
    wcss.append(k_module.inertia_)
    
```

```python
wcss
```

```python
## Elbow Curve

from matplotlib import pyplot as plt

def draw_plot()
plt.plot(range(1,11),wcss)
plt.title("Elbow Curve")
plt.xlabel("Clustomer_No")
plt.ylabel("WCSS")
plt.show
```

### Sillhouty score

```python
from sklearn.metrics import silhouette_score,calinski_harabasz_score
```

```python
sill_score = []
calinski_score = []
for i in range(2,11):
    k_module = KMeans(n_clusters = i,random_state = 12)
    predict_X = k_module.fit_predict(scalled_X)
    sill_Val = silhouette_score(scalled_X,predict_X)
    cal_Val = calinski_harabasz_score(scalled_X,predict_X)
    sill_score.append(sill_Val)
    calinski_score.append(sill_Val)
```

```python
sill_score,calinski_score
```

```python
def draw_plot(y,lable):
    plt.plot(range(2,11),y)
    plt.title(lable)
    plt.xlabel("Clustomer_No")
    plt.ylabel("Score")
    plt.show

print(draw_plot(sill_score,"SILL_Score_Plot"))
```

```python
print(draw_plot(calinski_score,"CaliNSKI_Score_Plot"))
```
