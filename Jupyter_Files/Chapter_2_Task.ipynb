```python
####### Create Data From with Timestamp Index 
import pandas as pd
import zoneinfo
from datetime import datetime,timezone

df = pd.DataFrame({
    "Order_Dt":["01/07/2025","01/07/2025","01/07/2025"],
    "sensor_id": ["TL-01", "TL-02", "TL-03"],
    "temperature": [23.5, 24.2, 22.8],
    "humidity": [45, 42, 47]
})

##Sequence : Get Zone detail --> timestamp --> timedeltawith NanoS --> Convert to int64 type
### Insert timestamp
##Get the Time Zone
start_time = zoneinfo.ZoneInfo("Asia/Riyadh")
## Get Current Time from the Zone
base_time = datetime.now(start_time)

## Make the Time Column Unify
df['time'] = [pd.Timestamp(base_time)+pd.Timedelta(nanoseconds=i) for i in range(len(df))]
df['time'] = pd.to_datetime(df['time']).astype('int64')
df.set_index('time',inplace = True)
df
```
