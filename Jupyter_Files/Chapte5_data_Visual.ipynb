```python
## 1. Histogram
```

```python
mt_cars.head()
```

```python
plt.hist(x = mt_cars.mpg)
plt.title("MPG Distribution")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```

```python
## Count Plat
```

```python
mt_cars["cyl"].value_counts()
```

```python
sns.countplot(data = mt_cars,x="cyl")
```

```python
### Pie Chart(Get the value count distribution of Decrete Data)
```

```python
mt_cars["carb"].value_counts()
```

```python
plt.figure(figsize = (7,7))
plt.pie(mt_cars["carb"].value_counts(),labels = ["4C","2C","1C","3C","6C","8C"]
       ,explode = [0.01,0.01,0.01,0.01,0.01,0.01],autopct = '%1.0f%%',pctdistance=0.7)

plt.title("Carb Distribution")
plt.show()
```

```python
## Heatmap

mt_cars_hm = mt_cars.copy()
mt_cars_hm.head(2)
```

```python
del mt_cars_hm["model"]
```

```python
mt_cars_hm.head(2)
```

```python

### Head MAp sutable for the correlation model
mt_cars_hm.corr()
```

```python
plt.figure(figsize= (10,10))
sns.heatmap(mt_cars_hm.corr(),annot=True)
```

## Notes(Correlation should be 0-1

 * 0 to 0.20 Very Weak Correlation
 * 0.21 to 0.40 Weak Correlation
 * 0.41 to 0.60 Normal Correlation
 * 0.61 - 0.80 Strong Correlation
 * 0.81 - 1.0 Very Strong Corelation
