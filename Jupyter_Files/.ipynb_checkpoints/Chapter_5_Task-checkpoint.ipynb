```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
penguin = sns.load_dataset("penguins")
```

```python
penguin.describe()
```

```python
penguin.isnull().sum()
```

```python
penguin[penguin["bill_length_mm"].isnull()]
```

```python
penguin.dropna(inplace = True)
```

```python
penguin.isnull().sum()
```

```python
penguin.dtypes
```

```python
penguin.head()
```

```python
## Univariant
plt.hist(penguin["bill_depth_mm"])
plt.xlabel("Bill Depth MM")
plt.ylabel("Penguin Population")
plt.title("Penguine Bill Depth")
plt.show()
```

```python
## Count Plot

sns.countplot(data = penguin,x= "sex",legend="auto")
plt.title("Penguien Gender Category")
plt.show()
```

```python
## Pie Chart

penguin["island"].value_counts()
```

```python
plt.pie(penguin["island"].value_counts(),explode= [0.005,0.005,0.005],autopct = '%1.2f%%',labels=["Biscoe","Dream","Torgerson"])
plt.title("Penguin Island")
plt.show()
```

```python
penguin.head(2)
```

```python
sns.barplot(penguin,x="species",y= "bill_depth_mm")
plt.title("Penguin Species Vs Bill Depth")
plt.show()
```

```python
## HeatMap
penguin_hm = penguin[["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]]
penguin_hm.corr()
```

```python
sns.heatmap(penguin_hm.corr())
plt.title("Penguin Metrix Correlation")
```
