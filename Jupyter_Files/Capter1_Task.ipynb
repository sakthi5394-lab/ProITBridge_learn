```python
### Objecttive

print(""" 
Build a simple Rock-Paper-Scissors game where the player competes with the computer.
 This task will test your understanding of:
●	Input/output
●	Conditional logic
●	Random module
●	Loops
●	Exception handling
""")

```

```python
### Instruction 

print("""
1.	The user should enter their choice: rock, paper, or scissors.
2.	The computer should randomly select one of these.
3.	Compare both and print who wins.
4.	Ask if the player wants to play again- if yes, repeat; if no, exit.
5.	Handle invalid inputs gracefully using try-except or condition checks.
""")
```

```python
import random

try:

    ## Create all the Required Valibles 
    game_choice = ('rock','paper','scissors')
    short_choice = ('r','p','s')
    total_try = 0
    com_w = 0
    usr_w = 0
    match_drw = 0
    invld_option = 0
    ## Loop Starts
    while True:
        total_try += 1
        ## Get the User Input
        computer_choice = random.choice(short_choice)
        user_choice_in = input(f" Enter your choice of {game_choice} : ").strip().lower()
        user_choice = user_choice_in[0]
        ## Conditiona Validation
        if user_choice == computer_choice:
            print("Match Draw \N{grinning face with smiling eyes}")
            match_drw += 1
            
        elif (user_choice not in game_choice) and (user_choice not in short_choice):
            print("Please enter Valid Option")
            invld_option += 1
            
        ## Validate the input with short value  r -- rock, p -- paper , s -scissors
        elif ((user_choice in ('r') and computer_choice == 's') or
              (user_choice == 'p' and computer_choice == 'r') or 
              (user_choice == 's' and computer_choice == 'p')):
                    print("Congrats! You win \N{party popper}")     
                    usr_w +=1
            
        else:
            print("Computer Wins \N{face with tears of joy}")
            com_w += 1

        ## Check user want to Play Again
        play_again = input("would you like to continue (Y/N)").strip().lower()
        if play_again == 'n':
            print(f"Total Tryies : {total_try}, You Win : {usr_w}, Computer_Win : {com_w}, Draw : {match_drw}, Invalid try : {invld_option}")
            print("Wining Percentage : ",round(int(usr_w)/int(total_try)*100,2),'%')
            break
        else:
            continue
## Print any Excpetion      
except Exception as e:
    print(f"Invald Option, Please retry \N{grinning face with smiling eyes} {e}")
```

## Word Guessing Game (Mini Hangman)

```python
print("""
Objective:
Create a word-guessing game where the player tries to guess the letters of a hidden word.

This will test your:
●	Loops
●	Conditional logic
●	String manipulation
●	Exception handling
●	Logical thinking

Instructions:
1.	Create a list of random words (like “python”, “machine”, “learning”, “bridge”).
2.	Randomly select one word for the game.
3.	Show the word as blanks (_ _ _ _) and let the player guess one letter at a time.
4.	Keep track of wrong guesses (max 6 attempts).
5.	End the game when the player either guesses the full word or uses all chances.


""")
```

```python
import random

try:
    word_list = ["python", "machine", "learning", "bridge"]
    secret_word = random.choice(word_list)
    ##print(secret_word)
    allowed_tries = len(secret_word)+3
    guessed_word = ["_"]*len(secret_word)
    total_try = 0

    print(f"Guess the {len(secret_word)} letter secret word  in {len(secret_word)+3} attemptes")
    
    ### Start the Loop
    while total_try < allowed_tries:
        total_try += 1
        user_input = input("Guess the charcter from the Secret Word:").strip().lower()
        ## Validate the User input having only one charcter
        if len(user_input) > 1 or user_input.isdigit():
            print("Enter only one character or valid character")
            continue

            
        ## If Guees letter Valid, Valid and Insert into Guessed List
        if user_input in guessed_word:
            print("Already Guesssed this character, Please try Another! Best of Luck")
            continue
        elif user_input in secret_word:
            for i in range(0,len(secret_word)):
                if user_input == secret_word[i] and guessed_word[i] == '_':
                    guessed_word[i] = user_input
        else:
            print(f"Select Another Charater which start with {secret_word[0]}")
            continue

        ## check word found or not 
        if secret_word == ''.join(guessed_word):
            print(f"Congrats! You find the Secret Word {secret_word} in {total_try} tries")
            break

        else:
            print(f"Your guessed word is {' '.join(guessed_word)}, remaining tries are {allowed_tries - total_try}")
            continue

        

    if '_' in guessed_word:
        print(f"Sorry All the {total_try} is over, Better Luck Next time")
   
except Exception as e:
    print(f"{e}")
```
